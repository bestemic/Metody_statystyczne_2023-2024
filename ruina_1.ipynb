# Ruina gracza
Proszę zasymulować proces ruiny gracza z kapitałami początkowymi:
a = 50,  b = 50.

Z symulacji dla różnych wartości p (prawdopodobieństwo zabrania przez gracza A 1$ od gracza B) proszę narysować wykresy zależności prawdopodobieństwa ruiny gracza P(ruina A) od p.
Powinny to być co najmniej dwa wykresy uśrednione po różnych liczbach gier, małej i dużej (np. 10 i 1000).


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def single_game(a_fortune, b_fortune, p):
    while a_fortune > 0 and b_fortune > 0:
        a_status = np.random.uniform(0, 1) < p
        a_fortune += 1 if a_status else -1
        b_fortune += -1 if a_status else 1 
    return a_fortune > 0
```

```python
def simulation(game_num, a_fortune, b_fortune, p):
    wins = 0
    for i in range(game_num):
        wins += single_game(a_fortune, b_fortune, p)
    return (game_num - wins) / game_num
```

```python
def calc_avg_probs(game_num, a_init_fortune, b_init_fortune, probs):
    avg_probs= []
    for p in probs:
        avg_probs.append(simulation(game_num, a_init_fortune, b_init_fortune, p))
    return avg_probs
```

```python
a_init_fortune = 50
b_init_fortune = 50
```

```python
game_num = 10
probs_part = np.arange(0.49, 0.52, 0.001)
avg_probs_part = calc_avg_probs(game_num, a_init_fortune, b_init_fortune, probs_part)
```

```python
probs = np.arange(0, 1, 0.02)
avg_probs = calc_avg_probs(game_num, a_init_fortune, b_init_fortune, probs)
```

```python
plt.figure(figsize=(12, 6))
plt.subplot(121)
plt.scatter(probs_part, avg_probs_part)
plt.xlabel('p')
plt.ylabel('Probability player A ruin')
plt.title(f'({game_num} games)')

plt.subplot(122)
plt.scatter(probs, avg_probs)
plt.xlabel('p')
plt.ylabel('Probability player A ruin')
plt.title(f'({game_num} games)')

plt.tight_layout()
plt.show()
```
