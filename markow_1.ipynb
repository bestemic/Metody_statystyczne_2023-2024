```python
import numpy as np
import matplotlib.pyplot as plt
import copy
```

```python
P = [[0.64, 0.32, 0.04],
     [0.4, 0.5, 0.1],
     [0.25, 0.5, 0.25]]
```

```python
P_new = copy.deepcopy(P)

tolerance = 1e-5
n = 0
norms = []

while True:
    n += 1
    
    P_old = copy.deepcopy(P_new)
    P_new = np.dot(P_old, P)
    
    norm = np.linalg.norm(P_new - P_old, ord='fro')
    norms.append(norm)

    if norm < tolerance:
        break

print("Stan stacjonarny Pi:", P_new[0])
print("Wartość n:", n)
```

### Norma Frobeniusa

$$
\|A\|_F = \sqrt{\sum_{i=1}^{m}\sum_{j=1}^{n} |a_{ij}|^2}
$$

```python
plt.plot(range(1, n + 1), norms)
plt.xlabel('Iteracje')
plt.ylabel('Norma')
plt.grid(True)
plt.show()
```

```python
Pi = np.array([1, 0, 0])

tolerance = 1e-5
n = 0
norms = []

while True:
    n += 1

    new_Pi = np.dot(Pi, P)
    
    norm = np.linalg.norm(new_Pi - Pi)
    norms.append(norm)

    if norm < tolerance:
        break

    Pi = new_Pi

print("Stan stacjonarny Pi:", Pi)
print("Wartość n:", n)
```

```python
plt.plot(range(1, n + 1), norms)
plt.xlabel('Iteracje')
plt.ylabel('Norma')
plt.grid(True)
plt.show()
```
